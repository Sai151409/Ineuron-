```python
# 1. Write a Python Program to Find the Factorial of a Number?

def factorial():
    """Its find out the factorial"""
    try :
        n = int(input("Enter the number : "))
        if n < 0:
            print("Enter positive number")
        elif n == 0:
            print(1)
        else:
            factorial = 1
        
            for i in range(1, n+1):
                factorial = factorial * i
            
        print(f"The factorial of {n} is {factorial}")
    except Exception as e:
        print("The Error has occured")
        print(f"The error is {e}")
```

```python
factorial()
```

```python
# 2. Write a Python Program to Display the multiplication Table?

# sol:

def table():
    """It give the multiplication table of given number"""
    try :
        n = int(input("Enter the table number that you want : "))
    
        print(f"The multiplication table of {n} : ")
    
        for i in range(1,11):
            print("\t\t", end = " ")
            print(f'{n} * {i} = {n*i}')
    except Exception as e:
        print("The Error has occured")
        print(f"The error is {e}")
```

```python
table()
```

```python
# 3. Write a Python Program to Print the Fibonacci sequence?

def fib():
    try:
        "Its gives the fibonacci series"
        n = int(input("Enter the number : "))
        a = 0
        b = 1
        for i in range(n):
            print(a)
            a, b = b, a+b
    except Exception as e:
        print("The Error has occured")
        print(f"The Error is {e}")
```

```python
fib()
```

```python
# 4. Write a Python Program to Check Armstrong Number?

def armstrong():
    "Its check whether the given number is armstrogn or not"
    try:
        n = int(input("Enter the number : "))
        temp = n
        sum = 0
        l = len(str(n))
        for i in range(len(str(n))):
            remainder = temp % 10
            sum = sum + remainder ** l
            temp = temp // 10
            
        if n == sum:
            print(f"{n} is armstrong number")
        else:
            print(f"{n} is not a armstrong number")
    except Exception as e:
        print("The Error has occured")
        print(f"The error is {e}")
```
